```python
import geopandas as gpd
```

```python
from zipfile import ZipFile

data = "data.zip"

with ZipFile(data, 'r') as zip:
    zip.printdir()
    zip.extractall() 
```

```python
areas_urbanas = gpd.read_file("/content/data/Areas Urbanas/areas_urbanas.shp")
distritos_censales = gpd.read_file("/content/data/Distritos Censales/Distritos Censales RM.shp")
estaciones = gpd.read_file("/content/data/Metro 2020/Estaciones_2020/Estaciones_2020.shp")
lineas = gpd.read_file("/content/data/Metro 2020/Lineas_2020/Lineas_2020.shp")
```

```python
print("Areas Urbanas")
display(areas_urbanas.dtypes)
display(areas_urbanas.head(10))
print("Distritos Censales")
display(distritos_censales.dtypes)
display(distritos_censales.head(10))
print("Estaciones")
display(estaciones.dtypes)
display(estaciones.head(10))
print("Lineas")
display(lineas.dtypes)
display(lineas.head(10))
```

```python
rm = distritos_censales.loc[distritos_censales["NOM_REG"] == "Región Metropolitana De Santiago"]
display(rm.head(10))
```

```python
distritos_y_areas_rm = gpd.overlay(rm, areas_urbanas, how= "intersection")
display(distritos_y_areas_rm.head(10))
```

```python
distritos_y_areas_santiago = distritos_y_areas_rm.loc[distritos_y_areas_rm["NOM_PROV"]== "Santiago"]
display(distritos_y_areas_santiago.head(10))
```

```python
distritos_y_areas_santiago.plot()
```

```python
import osmnx as ox
```

```python
ox.config(log_console=True, use_cache=True)

places = ["Macul, Santiago, Chile"]

stats = {}
for place in places:
    G = ox.graph_from_place(place, network_type='drive')
    stats[place] = ox.basic_stats(G)
display(gpd.GeoDataFrame(stats).T)
ox.plot_graph(G)
```
